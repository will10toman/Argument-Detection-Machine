```python
import pandas as pd
import os

def add_to_argument_detection(dataset, column, label, output_path="argumentdetection6.csv"):
    """
    Adds data from any dataframe (whatever its name) and column to a CSV with two columns: sentence, category.
    You pass the dataframe variable, column name, and category label each time you call it.
    """

    if column not in dataset.columns:
        raise ValueError(f"Column '{column}' not found in dataset. Available columns: {list(dataset.columns)}")


    new = pd.DataFrame({
        "sentence": dataset[column].astype(str),
        "category": label
    })


    if not os.path.exists(output_path):
        new.to_csv(output_path, index=False, encoding="utf-8")
        print(f"Created new {output_path} with {len(new)} rows.")
    else:
        existing = pd.read_csv(output_path)
        if not set(["sentence", "category"]).issubset(existing.columns):
            # If the old CSV had weird columns, fix it
            existing = existing.rename(columns={existing.columns[0]: "sentence", existing.columns[1]: "category"})
            existing = existing[["sentence", "category"]]
        
        combined = pd.concat([existing, new], ignore_index=True)
        combined.to_csv(output_path, index=False, encoding="utf-8")
        print(f"Appended {len(new)} rows â†’ total now {len(combined)} rows in {output_path}")
```
